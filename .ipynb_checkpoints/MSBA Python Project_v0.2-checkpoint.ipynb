### Naive's Bayes Implementation

```python
##### Setup for classification #####
import re
#Only analysing the Right trolls

mask1 = troll_df["account_category"] == "RightTroll"
troll_acct_df = troll_df[mask1]

#Get words

def make_it_into_words_list(new_quote):
    # Input: A full comment
    # Output: a list of words in the comment
    x = str(new_quote)
    x = x.lower()
    x.split("(\s)")
    x=re.findall("[A-Za-z0-9_\`]+", x)
    return x

#Count occurence of words

def AcctwordsList(s):
    aux = make_it_into_words_list(s)
    for word in aux:
        if word in AcctUniqueWords:
            AcctUniqueWords[word] = AcctUniqueWords[word] + 1
        else:
            AcctUniqueWords[word] = 1
        
    return AcctUniqueWords
AcctUniqueWords = {}
```

```python
AcctUniqueWords=troll_acct_df["content"].map(AcctwordsList)
```

```python
AcctWords= Series(AcctUniqueWords[1], index = AcctUniqueWords[1].keys())
AcctWords[:5]
```

```python
AQ = AcctWords.sort_values(ascending = False )
AQ[:40]
```

```python
Right500Words = AQ[:5000].index.tolist()
Right500Words[:5]
```

```python
#Get Hasttags

def accthashtag(s):
    A = str(s)
    X = re.findall(" \#([A-Za-z]+)",A)
    if X != []:
        for word in X:
            if word in AcctHashtags:
                AcctHashtags[word] = AcctHashtags[word] + 1
            else:
                AcctHashtags[word] = 1
        return AcctHashtags

AcctHashtags = {}
Accthashtags = troll_acct_df["content"].map(accthashtag)
```

```python
AcctHashtags= Series(AcctHashtags, index = AcctHashtags.keys())
HAcct = AcctHashtags.sort_values(ascending = False )
Right50Hash = HAcct[:500].index.tolist()

def Tweetlength(s):
    return len(str(s).split(" "))
def InTopHashtags(s):
    A = str(s)
    X = re.findall(" \#([A-Za-z]+)",A)
    count = 0
    if X != []:
        for word in X:
            if word in Right50Hash:
                count += 1
    return count
def InTopWords(s):
    A = str(s)
    X = re.findall(" \#([A-Za-z]+)",A)
    count = 0
    if X != []:
        for word in X:
            if word in Right500Words:
                count += 1
    return count

# Train new data frame
tdf2 = pd.DataFrame()
tdf2["region"] = train_df["region"]
tdf2["language"] = train_df["language"]
tdf2["account_category"] = train_df["account_category"]
tdf2["account_type"] = train_df["account_type"]
tdf2["content"] = train_df["content"]
tdf2["retweet"] = train_df["retweet"]
tdf2["regionUS"] = 0
tdf2["regionUS"][tdf2["region"] == "United States"] = 1
tdf2["English"] = 0
tdf2["English"][tdf2["language"] == "English"] = 1
tdf2["righttroll"] = 0
tdf2["righttroll"][tdf2["account_category"] == "RightTroll"] = 1
tdf2["NumTopWords"] = tdf2["content"].map(InTopWords)
tdf2["NumTopHash"] = tdf2["content"].map(InTopHashtags)
tdf2["Tweet Length"] = tdf2["content"].map(Tweetlength)

# Test new data frame
tdf3 = pd.DataFrame()
tdf3["region"] = test_df["region"]
tdf3["language"] = test_df["language"]
tdf3["account_category"] = test_df["account_category"]
tdf3["account_type"] = test_df["account_type"]
tdf3["content"] = test_df["content"]
tdf3["retweet"] = test_df["retweet"]
tdf3["regionUS"] = 0
tdf3["regionUS"][tdf3["region"] == "United States"] = 1
tdf3["English"] = 0
tdf3["English"][tdf3["language"] == "English"] = 1
tdf3["righttroll"] = 0
tdf3["righttroll"][tdf3["account_category"] == "RightTroll"] = 1
tdf3["NumTopWords"] = tdf3["content"].map(InTopWords)
tdf3["NumTopHash"] = tdf3["content"].map(InTopHashtags)
tdf3["Tweet Length"] = tdf3["content"].map(Tweetlength)
```

```python
#Naive's Bayes Implementation


from patsy import dmatrices
from sklearn import naive_bayes
from sklearn import metrics
tdf2["righttroll"].value_counts()


Y,X = dmatrices('righttroll ~ 0 + tdf2["retweet"]+tdf2["English"]+tdf2["regionUS"]+tdf2["NumTopWords"]+tdf2["NumTopHash"]'
, tdf2, return_type = 'dataframe')

y = Y["righttroll"].values
model = naive_bayes.MultinomialNB()
model.fit(X,Y)

print 'Prediction'
print model.predict(X[:10])
print 'Actual'
print y[:10]

prediction_train = model.predict(X)
print metrics.accuracy_score(y, prediction_train)

Y_test, X_test = dmatrices('righttroll ~ 0 + tdf3["retweet"]+tdf3["English"]+tdf3["regionUS"]+tdf3["NumTopWords"]+tdf3["NumTopHash"]'
, tdf3, return_type='dataframe')

y_test = Y_test['righttroll'].values
prediction_test = model.predict(X_test)
print metrics.accuracy_score(y_test, prediction_test)
print "pred 0 1"
print metrics.confusion_matrix(y_test, prediction_test, labels = [0,1])
```
